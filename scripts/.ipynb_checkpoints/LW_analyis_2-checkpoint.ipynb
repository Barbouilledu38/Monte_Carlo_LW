# Compréhension des coefficients d'absorption par l'atm

Pour chaque bande de $\lambda$, les coefficient d'absorption $k_a(\lambda)$ sont rangés par ordre croissant $k_a(g)$. La nouvelle distribution $k_a(g)$ est alors monotone croissante est couvre plusieurs ordre de grandeur. On peut intégrer numériquement cette fonction simplement en prenant quelque g-points : $$k_{\Delta\lambda_j,eff}(\Delta z_i) = \sum_g w_{g,\Delta\lambda_j} k(g,\Delta z_i,\Delta\lambda_j)$$
Avec $\Delta\lambda_j$, $j \in [0,15]$, les bandes de fréquences décrivant l'ensemble du spectre thermique. $\Delta z_i$, $i \in [0,30]$, les couches atmosphérique 1D. $w_g$ les poids de chaque point de quadrature (g-point) pour chaque bande, pour chaque bande.

```python
fig, axs = plt.subplots(nrows = 2, ncols = 2, figsize=(20, 9), layout = 'constrained')
save_dir = "/home/barroisl/Transect_MC_auto/Output/"
for i in range(4):
    ax = axs[i//2,i%2]
    atm = atms[i]
    
    if atm != "EMPTATM" and atm != "TUXUATM":
        
        prop_atm = Prop_atm(atm)
        
        if i//2 == 0 :
            up_label = True
            x_lab = False
        else :
            up_label = False
            x_lab = True

        if i%2 == 0 :
            y_lab = True
        else :
            y_lab = False

        if i == 1:
            colorbar = True
        else :
            colorbar = False

        prop_atm.plotting_2D_ka(ax = ax, axs = axs, up_label = up_label, colorbar = colorbar, \
                                force_save = False,save_name = save_dir + f"ka_profile_2D_{atm}.png",\
                               y_lab = y_lab, x_lab = x_lab)

    elif atm == "EMPTATM":
        
        prop_atm = Prop_atm("SMLSATM")
        
        alt_km = prop_atm.profiles(variable = "z")*1e-3
        bands = prop_atm.LW_int_edges
        arr_nb_z = prop_atm.tab_k_z_nb()

        im = ax.pcolormesh(
            np.arange(1, arr_nb_z.shape[1] + 1),  
            alt_km,                                  
            np.full_like(arr_nb_z,0),                             
            norm=LogNorm(vmin=1e-7, vmax=1e0),
            cmap='binary',
            shading='auto'
        )
        
        ax.set_xticks([])
        ax.set_label('Altitude [km]')
        
        ax2 = ax.twiny()
        ax2.set_xlim(ax.get_xlim())
        ax2.set_xticks(range(1, arr_nb_z.shape[1] + 1))
        ax2.set_xticklabels([f"{b[0]}-{b[1]} µm" for b in bands], fontsize=15, rotation=45, ha='left')
        #ax2.set_xlabel('Intervalle [cm⁻¹]')

    elif atm == "TUXUATM":
        
        prop_atm = Prop_atm("SMLSATM")
        
        alt_km = prop_atm.profiles(variable = "z")*1e-3
        bands = prop_atm.LW_int_edges
        arr_nb_z = prop_atm.tab_k_z_nb()

        im = ax.pcolormesh(
            np.arange(1, arr_nb_z.shape[1] + 1),  
            alt_km,                                  
            np.full_like(arr_nb_z,0.002),                             
            norm=LogNorm(vmin=1e-7, vmax=1e0),
            cmap='binary',
            shading='auto'
        )
        ax.set_xlabel('Bande spectrale LW')
        ax.set_xticks(range(1, arr_nb_z.shape[1] + 1))
        ax.set_yticks([])
        
    ax.set_title(dict_atm[atm][0],fontsize = 20)
    
plt.savefig(save_dir+'comparaison_ka_profile.png')
```

# Nouvelle approche flux atmosphérique

Afin de se rapprocher du but de la manoeuvre (la descente d'échelle), on cherche à prendre en compte la partie des chemins émis par l'espace par un terme d'émissivité

On fait comme si l'émission de l'atmosphère était celle d'une surface type corps noir à une certaine température (possible celle à 1 km dans l'atmosphère). L'émissivité de ce corps noir est calculée comme le rapport entre ... 

```python
def z_profile_LWd(
    dir_data : str = "/home/barroisl/Transect_MC_auto/Data/z_profile_atms/"):
    
    atms = ["SMLSATM","SMLWATM","TUXUATM"]
    
    dtype = np.dtype([
        ("z",    np.float64),
        ("SMLSATM",     np.float64),
        ("SMLWATM",     np.float64),
        ("TUXUATM",     np.float64),
    ])
    
    n_dist = len(os.listdir(dir_data + "SMLSATM/"))
    print(n_dist)
    flux_d_atm = np.zeros(n_dist, dtype=dtype)
    
    for j,atm in enumerate(atms) :
        
        list_dir = os.listdir(dir_data + atm + "/")
        print(len(list_dir))
        
        z_list = []
        fd_list = []

        for i,dossier in tqdm(enumerate(list_dir)):

            # Data des chemins htrdr
            data_path = np.loadtxt(dir_data + atm +"/" + dossier + '/merge' +'_50_15_15.txt')
            pos_camera = np.array([240150,5001981,float(dossier)])
            mc_set = MC_Set(data_path, pos_camera)

            z_list.append(float(dossier))
            fd_list.append(mc_set.W_tot/mc_set.Nbre_photon)
            
        print(len(fd_list))    
            
        flux_d_atm[atm] = fd_list

        if j == 0 :
            flux_d_atm["z"] = z_list
    
    np.save(dir_data +"flux_d.npy", flux_d_atm)
```

```python
profile_LWd = z_profile_LWd()
```

```python
flux_d = np.load("/home/barroisl/Transect_MC_auto/Data/z_profile_atms/flux_d.npy")

fig, ax = plt.subplots(figsize = (12,6), layout = 'constrained')

ax2 = ax.twinx()

for atm in atms :
    if atm != "EMPTATM" and atm !="TUXUATM":

        prop_atm = Prop_atm(atm)
        T_profile = prop_atm.profiles(variable = "T")
        z_profile = prop_atm.profiles(variable = "z") 
        
        ax2.plot(sigma*T_profile**4, z_profile, color = dict_atm[atm][1],\
            linestyle = 'dashed')

        
        flux_d_atm = np.interp(z_profile, flux_d["z"][1:], flux_d[atm][1:])
        ax.plot(flux_d_atm,z_profile,label = dict_atm[atm][0], color = dict_atm[atm][1],\
                marker = 'o')

    ax2.plot(sigma*(298+(-6.5e-3)*(z_profile[1:]-0))**4, z_profile[1:], color = "purple",\
        marker = 'o')


ax2.set_ylim((0,4100))
ax2.spines[['left','right', 'top', 'bottom']].set_visible(False)
ax2.set_ylabel(' ')

ax.set_ylim((0,4100))
ax.spines[['left','right', 'top', 'bottom']].set_visible(False)
ax.set_xlabel('LWd [$W.m^{-2}$]')
ax.set_ylabel('Altitude [$m$]')
ax.legend()
```
